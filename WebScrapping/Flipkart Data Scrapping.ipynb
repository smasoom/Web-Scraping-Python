```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
company= soup.find_all('div', class_='yKfJKb row')
```

```python
ratings = []

for item in company:
    rating_element = item.find('div', class_='XQDdHH')
    if rating_element:
        rating_text = rating_element.text.strip() 
```

```python
processor = []
price = []
moredata = []
nofrating = []

base_url = 'https://www.flipkart.com/search?q=laptop%20under%201lakh&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='

for page_num in range(1, 5):  
    url = base_url + str(page_num)
    webpage = requests.get(url, headers=headers).text
    company = soup.find_all('div', class_='yKfJKb row')
    
    for item in company:
        about_element = item.find('div', class_='KzDlHZ')
        rating_element = item.find('div', class_='XQDdHH')
        Processor = item.find('li', class_='J+igdf')
        Moredata = item.find('ul', class_='G4BRas')
        Price = item.find('div', class_='Nx9bqj _4b5DiR')
        Nofrating = item.find('span', class_='Wphh3N')

        if about_element:
            about.append(about_element.text)
        else:
            about.append('N/A')
            
        if rating_element:
            rating.append(rating_element.text)
        else:
            rating.append('N/A')
            
        if Processor:
            processor.append(Processor.text)
        else:
            processor.append('N/A')
            
        if Price:
            price.append(Price.text)
        else:
            price.append('N/A')
            
        if Moredata:
            moredata.append(Moredata.text)
        else:
            moredata.append('N/A')
            
        if Nofrating:
            nofrating.append(Nofrating.text)
        else:
            nofrating.append('N/A')

print("Processor:", processor)
print("Price:", price)
print("More Data:", moredata)
print("Number of Ratings:", nofrating)
```
